```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("UNRATE.csv")
all_y = np.array(df.UNRATE[5:][::12].values)
```

```python
all_x = range(1948, 2021)
plt.style.use('ggplot')
plt.plot(all_x, all_y)
plt.title("US unemployment rate vs Time")
plt.ylabel("Us unemployment rate")
plt.xlabel("Time (year)")
plt.savefig("all.png", bbox_inches="tight")
```

```python
this_year_y = np.array(df.UNRATE[-6:].values)
this_year_x = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
placeholder_x = range(0, 6)
```

```python
plt.plot(placeholder_x, this_year_y)
plt.xticks(ticks=placeholder_x, labels=this_year_x)
plt.title("US unemployment rate (2020) vs Time")
plt.ylabel("Us unemployment rate (2020)")
plt.xlabel("Time (month)")
plt.savefig('2020.png', bbox_inches='tight')
```
